```python
from tkinter import *
from tkinter import filedialog
from tkinter import font
from tkinter import colorchooser

#setting a variable for open file name
global open_stat_name
open_stat_name = False

#creating function for new file
def new_file():
    #Delete previous text
    textbox.delete(1.0, END)
    
    #updating the new file
    root.title("New File - Text Editor!")
    
    global open_stat_name
    open_stat_name = False
    
#creating function for open file
def open_file():
    #Delete previous text
    textbox.delete(1.0, END)
    
    #Grab filename
    text_file=filedialog.askopenfilename(initialdir="C:/Data Science/",
                                         title="Open File",
                                         filetypes=(("Text Files","*.txt"),
                                                    ("HTML Files","*.html"),
                                                    ("Python Files","*.py"),
                                                    ("All Files","*.*")))
    # to check if a file already exist
    if text_file:
        #make file global to access it later
        global open_stat_name
        open_stat_name = text_file
    
    #update status bar
    name = text_file
    name = name.replace("C:/Users/user/Desktop/Data Science/", "")
    root.title(f'{name} - Text Editor!')
    
    #open the file
    text_file = open(text_file, 'r')
    anime=text_file.read()
    #add file to textbox
    textbox.insert(END, anime)
    #close the opened file
    text_file.close()

#creating function for save as file
def save_as_file():
    text_file = filedialog.asksaveasfilename(defaultextension=".*",
                                             initialdir="C:/Data Science/",
                                             title="Save File",
                                             filetypes=(("Text File","*.txt"),
                                                        ("HTML Files","*.html"),
                                                        ("Python Files","*.py"),
                                                        ("All Files","*.*")))
    if text_file:
        #update status bar
        name = text_file
        name = name.replace("C:/Users/user/Desktop/Data Science/", "")
        root.title(f'{name} - Text Editor!')
        
    # save the file
    text_file = open(text_file, 'w')
    text_file.write(textbox.get(1.0, END))
    #close the file
    text_file.close()

#creatin a function for save
def save_file():
    global open_stat_name
    if open_stat_name:
        # save the file
        text_file = open(open_stat_name, 'w')
        text_file.write(textbox.get(1.0, END))
        #close the file
        text_file.close()
    else:
        save_as_file()
        
#creating a function for bold
def bold():
    #create our font
    bold_text=font.Font(textbox,textbox.cget("font"))
    bold_text.configure(weight="bold")
    
    #configuring a tag
    textbox.tag_configure("bold", font=bold_text)
    
    #defining a tag
    current_tags=textbox.tag_names("sel.first")
    
    #to check if something has been made bold or not
    if "bold" in current_tags:
        textbox.tag_remove("bold", "sel.first", "sel.last")
    else:
        textbox.tag_add("bold", "sel.first", "sel.last")

#creating function for italics
def italic():
    #create our font
    italic_text=font.Font(textbox,textbox.cget("font"))
    italic_text.configure(slant="italic")
    
    #configuring a tag
    textbox.tag_configure("italic", font=italic_text)
    
    #defining a tag
    current_tags=textbox.tag_names("sel.first")
    
    #to check if something has been made bold or not
    if "italic" in current_tags:
        textbox.tag_remove("italic", "sel.first", "sel.last")
    else:
        textbox.tag_add("italic", "sel.first", "sel.last")

def font_colour():
    #pick a color
    my_color = colorchooser.askcolor()[1]
    if my_color:
        #create our font
        color_font=font.Font(textbox,textbox.cget("font"))
    
        #configuring a tag
        textbox.tag_configure("colored", font=color_font, foreground=my_color)
    
        #defining a tag
        current_tags=textbox.tag_names("sel.first")
    
        #to check if text color has changed or not
        if "colored" in current_tags:
            textbox.tag_remove("colored", "sel.first", "sel.last")
        else:
            textbox.tag_add("colored", "sel.first", "sel.last")
            
def font_underline():
    #create our font
    underline_font=font.Font(textbox,textbox.cget("font"))
    
    #configuring a tag
    textbox.tag_configure("undelined", font=underline_font,underline= True )
    
    #defining a tag
    current_tags=textbox.tag_names("sel.first")
    
    #to check if text color has changed or not
    if "underlined" in current_tags:
        textbox.tag_remove("underlined", "sel.first", "sel.last")
    else:
        textbox.tag_add("underlined", "sel.first", "sel.last")
    

root=Tk(className=" Text Editor")

#creating a frame for the buttons
mar=Frame(root)
mar.grid(row=0,column=0,)

#creating the main frame
mark=Frame(root)
mark.grid(row=1,column=0,pady=8)

#creating toolbar buttons
#Bold button
bold=Button(mar,text="Bold",command=bold)
bold.grid(row=0,column=0)

#Italic Button
ital=Button(mar,text="Italics",command=italic)
ital.grid(row=0,column=1,padx=5)

#underline button
und=Button(mar,text="Underline", command=font_underline)
und.grid(row=0,column=2)

#underline button
fntsy=Button(mar,text="Font Style")
fntsy.grid(row=0,column=3,padx=5)

# Font colour button
fntcl=Button(mar,text="Font Colour", command=font_colour)
fntcl.grid(row=0,column=4)

#bullet button
bullet=Button(mar,text="Bullet")
bullet.grid(row=0,column=5,padx=5)

#creating scrollbar for the textbox
text_scrollbar = Scrollbar(mark)
text_scrollbar.grid(row=2,column=1)

#creating the text box
textbox=Text(mark,
             width=97,
             height=25,
             font=("TImes",16),
             selectbackground="pink",
             selectforeground="black",
             undo=True,
             yscrollcommand=text_scrollbar.set)
textbox.grid(row=2,column=0)

#configuring the scrollbar
text_scrollbar.config(command=textbox.yview)

#creating menu
menu_=Menu(root)
root.config(menu=menu_)

#adding file menu
file_menu= Menu(menu_, tearoff=False)
menu_.add_cascade(label="File",menu=file_menu)
file_menu.add_command(label="New",command=new_file)
file_menu.add_command(label="Open",command=open_file)
file_menu.add_command(label="Save",command=save_file)
file_menu.add_command(label="Save As",command=save_as_file)
file_menu.add_separator()
file_menu.add_command(label="Exist", command=root.quit)

#adding edit menu
edit_menu= Menu(menu_, tearoff=False)
menu_.add_cascade(label="Edit",menu=edit_menu)
edit_menu.add_command(label="Redo",command=textbox.edit_redo,accelerator="(Ctrl+y)")
edit_menu.add_command(label="undo",command=textbox.edit_undo,accelerator="(Ctrl+z)")
root.mainloop()
```
